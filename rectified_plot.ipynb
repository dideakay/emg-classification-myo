```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read the CSV file
df = pd.read_csv('fist_0.csv')
```

```python

# Extract the sensor columns
sensor_columns = df.columns
# Apply rectification to the data
df_rectified = df.abs()
plt.figure(figsize=(20, 15))

# Create a time column based on the assumption of 200 Hz frequency
time_values = df.index / 200.0  # Assuming data is acquired at 200 Hz

# Define legend names and colors
legend_names = ['Sensor 1', 'Sensor 2', 'Sensor 3', 'Sensor 4', 'Sensor 5', 'Sensor 6', 'Sensor 7', 'Sensor 8']
colors = ['red', 'blue', 'green', 'orange', 'purple', 'brown', 'gray', 'pink']



# Plot each sensor as a signal graph with custom legend names and colors
for i, column in enumerate(sensor_columns):
    line, = plt.plot(time_values, df_rectified[column], color=colors[i])  # Store the line object
    line.set_label(legend_names[i])
#
# Set plot labels and title
plt.xlabel('Time (s)')
plt.ylabel('Rectified Signal Value')
plt.title('Rectified Sensor Signals')
plt.xlim(0.5,2.5)
# Add legend
plt.legend(legend_names)

# Display the plot
plt.show()
```

```python
# Create separate subplots for each sensor
num_sensors = len(sensor_columns)
fig, axs = plt.subplots(num_sensors, 1, sharex=True, figsize=(20, 30))  # Adjust figsize as needed

# Plot each sensor data in separate subplots with custom legend names and colors
for i, column in enumerate(sensor_columns):
    axs[i].plot(time_values, df_rectified[column], label=legend_names[i], color=colors[i])
    axs[i].set_ylabel('Signal Value')
    axs[i].set_title(legend_names[i])

# Set common x-axis label and title
plt.xlabel('Time (s)')
fig.suptitle('Sensor Signals')

# Add legend to the last subplot
axs[num_sensors-1].legend()

# Adjust spacing between subplots
#plt.tight_layout()
plt.xlim(0.5, 2.5)

# Display the plot
plt.show()
```
